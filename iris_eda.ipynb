```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
#the above tells us that the data is loaded succesfully
```

```python
df.describe()
```

```python
df.mean(numeric_only=True)
```

```python
df.median(numeric_only=True)
```

```python
df.std(numeric_only=True)
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['sepal_length'], bins=20, color='tomato', edgecolor='black')
plt.title("1. Histogram of sepal_length")
plt.xlabel("sepal_length")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['sepal_width'], bins=20, color='cornflowerblue', edgecolor='black')
plt.title("2. Histogram of sepal_width")
plt.xlabel("sepal_width")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['petal_length'], bins=20, color='mediumseagreen', edgecolor='black')
plt.title("3. Histogram of petal_length")
plt.xlabel("petal_length")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
